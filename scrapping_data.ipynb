# **Scraping Dataset**

```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time

app_id = 'com.gojek.app'
total_reviews = 25000
batch_size = 100

all_reviews = []
next_token = None

while len(all_reviews) < total_reviews:
    r, next_token = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=batch_size,
        continuation_token=next_token
    )
    all_reviews.extend(r)

    if not next_token:
        break

    time.sleep(1)
```

# **Loading Dataset**

```python
df = pd.DataFrame(all_reviews)
print("Jumlah total data:", len(df))
df.head()
```

```python
df = df[['content', 'score']]
print("Jumlah total data:", len(df))
df.head()
```

```python
df.to_csv('gojek_reviews.csv', index=False, encoding='utf-8-sig')
print("Selesai! File disimpan sebagai `gojek_reviews.csv`")
```
